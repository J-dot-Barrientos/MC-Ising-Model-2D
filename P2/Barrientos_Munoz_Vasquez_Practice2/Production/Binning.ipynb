```python
import numpy as np
import matplotlib.pyplot as plt

# cargar datos
data = np.loadtxt("FULLCRITICAL.dat")

E = data[:, 0]
M = data[:, 1]

# tiempo real en MC steps
time = 10 * np.arange(len(E))

E = E[::100]
M = M[::100]
time = time[::100]
# -------------------------
# plots
# -------------------------

# E vs time
plt.figure()
plt.plot(time, E/10000)
plt.xlabel("Monte Carlo Step")
plt.ylabel("E/N")
plt.title("Energy evolution")
plt.tight_layout()
plt.savefig("Energy_critical_full.png", dpi=600)

# M vs time
plt.figure()
plt.plot(time, M/10000)
plt.xlabel("Monte Carlo Step")
plt.ylabel("M/N")
plt.title("Magnetization evolution")
plt.tight_layout()
plt.savefig("Magnetization_critical_full.png", dpi=600)

plt.show()
```
